## Polish-Czech

```python
X_train = pd.read_csv('polish_czech_train.csv')
```

```python
X_train
```

```python
from collections import defaultdict
# unigrams

unigrams_pol = defaultdict(int)
bigrams_pol = defaultdict(int)
trigrams_pol = defaultdict(int)
fourgrams_pol = defaultdict(int)

for idx, row in X_train[X_train['lan_code'] == 'pol'].iterrows():
    for i in range(len(row['sentence'])):
        if len(row['sentence'][i:i+1]) == 1:
            unigrams_pol[row['sentence'][i:i+1]] += 1
        
        if len(row['sentence'][i:i+2]) == 2:
            bigrams_pol[row['sentence'][i:i+2]] += 1
        
        if len(row['sentence'][i:i+3]) == 3:
            trigrams_pol[row['sentence'][i:i+3]] += 1
        
        if len(row['sentence'][i:i+4]) == 4:
            fourgrams_pol[row['sentence'][i:i+4]] += 1
        
        
    
```

```python
unigrams_ces = defaultdict(int)
bigrams_ces = defaultdict(int)
trigrams_ces = defaultdict(int)
fourgrams_ces = defaultdict(int)

for idx, row in X_train[X_train['lan_code'] == 'ces'].iterrows():
    for i in range(len(row['sentence'])):
        if len(row['sentence'][i:i+1]) == 1:
            unigrams_ces[row['sentence'][i:i+1]] += 1
        
        if len(row['sentence'][i:i+2]) == 2:
            bigrams_ces[row['sentence'][i:i+2]] += 1
        
        if len(row['sentence'][i:i+3]) == 3:
            trigrams_ces[row['sentence'][i:i+3]] += 1
        
        if len(row['sentence'][i:i+4]) == 4:
            fourgrams_ces[row['sentence'][i:i+4]] += 1
```

### Prediction

```python
X_test = pd.read_csv('polish_czech_test.csv')

X_test
```

```python
def predict(sentence, N, ngramDict_pol, ngramDict_ces):
    pol_count = 0
    ces_count = 0
    for i in range(len(sentence) - N + 1):
        pol_count += ngramDict_pol[sentence[i:i+N]]
        ces_count += ngramDict_ces[sentence[i:i+N]]
    
    if pol_count > ces_count:
        return 'pol'
    else:
        return 'ces'
```

```python
X_test['lan_code_unigrams'] = X_test['sentence'].apply(lambda s: predict(s, 1, unigrams_pol, unigrams_ces))
X_test['lan_code_bigrams'] = X_test['sentence'].apply(lambda s: predict(s, 2, bigrams_pol, bigrams_ces))
X_test['lan_code_trigrams'] = X_test['sentence'].apply(lambda s: predict(s, 3, trigrams_pol, trigrams_ces))
X_test['lan_code_fourgrams'] = X_test['sentence'].apply(lambda s: predict(s, 4, fourgrams_pol, fourgrams_ces))

X_test
```

```python
X_test.to_csv('prediction.csv', index=False)
```

```python
incorrect_predictions = X_test[X_test.lan_code_unigrams != X_test.lan_code]
incorrect_predictions = pd.concat([incorrect_predictions, X_test[X_test.lan_code_bigrams != X_test.lan_code]])
incorrect_predictions = pd.concat([incorrect_predictions, X_test[X_test.lan_code_trigrams != X_test.lan_code]])
incorrect_predictions = pd.concat([incorrect_predictions, X_test[X_test.lan_code_fourgrams != X_test.lan_code]])

incorrect_predictions = incorrect_predictions.drop_duplicates()

incorrect_predictions.to_csv('incorrect.csv', index=False)
```

```python
def zero_count_ngrams(df, N, ngramDict_pol, ngramDict_ces):
    pol_zero_count = set()
    ces_zero_count = set()
    
    for _, row in df.iterrows():
        lan_code = row['lan_code']
        sentence = row['sentence']
        
        for i in range(len(sentence) - N + 1):
            if lan_code == 'ces' and ngramDict_ces[sentence[i:i+N]] == 0:
                ces_zero_count.add(sentence[i:i+N])
            
            if lan_code == 'pol' and ngramDict_pol[sentence[i:i+N]] == 0:
                pol_zero_count.add(sentence[i:i+N])
        
    return pol_zero_count, ces_zero_count
                
```

```python
zc_uni_pol, zc_uni_ces = zero_count_ngrams(X_test, 1, unigrams_pol, unigrams_ces)

with open('zc_uni_pol.json', 'w') as f:
    json.dump(list(zc_uni_pol), f, indent=' ')

with open('zc_uni_ces.json', 'w') as f:
    json.dump(list(zc_uni_ces), f, indent=' ')
```

```python
zc_bi_pol, zc_bi_ces = zero_count_ngrams(X_test, 2, bigrams_pol, bigrams_ces)

with open('zc_bi_pol.json', 'w') as f:
    json.dump(list(zc_bi_pol), f, indent=' ')

with open('zc_bi_ces.json', 'w') as f:
    json.dump(list(zc_bi_ces), f, indent=' ')
```
