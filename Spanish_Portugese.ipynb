## Spanish-Portugese

```python
X_train = pd.read_csv('spanish_portugese_train.csv')
```

```python
X_train
```

```python
from collections import defaultdict
# unigrams

unigrams_spa = defaultdict(int)
bigrams_spa = defaultdict(int)
trigrams_spa = defaultdict(int)
fourgrams_spa = defaultdict(int)

for idx, row in X_train[X_train['lan_code'] == 'spa'].iterrows():
    for i in range(len(row['sentence'])):
        if len(row['sentence'][i:i+1]) == 1:
            unigrams_spa[row['sentence'][i:i+1]] += 1
        
        if len(row['sentence'][i:i+2]) == 2:
            bigrams_spa[row['sentence'][i:i+2]] += 1
        
        if len(row['sentence'][i:i+3]) == 3:
            trigrams_spa[row['sentence'][i:i+3]] += 1
        
        if len(row['sentence'][i:i+4]) == 4:
            fourgrams_spa[row['sentence'][i:i+4]] += 1
        
        
    
```

```python
unigrams_por = defaultdict(int)
bigrams_por = defaultdict(int)
trigrams_por = defaultdict(int)
fourgrams_por = defaultdict(int)

for idx, row in X_train[X_train['lan_code'] == 'por'].iterrows():
    for i in range(len(row['sentence'])):
        if len(row['sentence'][i:i+1]) == 1:
            unigrams_por[row['sentence'][i:i+1]] += 1
        
        if len(row['sentence'][i:i+2]) == 2:
            bigrams_por[row['sentence'][i:i+2]] += 1
        
        if len(row['sentence'][i:i+3]) == 3:
            trigrams_por[row['sentence'][i:i+3]] += 1
        
        if len(row['sentence'][i:i+4]) == 4:
            fourgrams_por[row['sentence'][i:i+4]] += 1
```

### Prediction

```python
X_test = pd.read_csv('spanish_portugese_test.csv')

X_test
```

```python
def predict(sentence, N, ngramDict_spa, ngramDict_por):
    spa_count = 0
    por_count = 0
    for i in range(len(sentence) - N + 1):
        spa_count += ngramDict_spa[sentence[i:i+N]]
        por_count += ngramDict_por[sentence[i:i+N]]
    
    if spa_count > por_count:
        return 'spa'
    else:
        return 'por'
```

```python
X_test['lan_code_unigrams'] = X_test['sentence'].apply(lambda s: predict(s, 1, unigrams_spa, unigrams_por))
X_test['lan_code_bigrams'] = X_test['sentence'].apply(lambda s: predict(s, 2, bigrams_spa, bigrams_por))
X_test['lan_code_trigrams'] = X_test['sentence'].apply(lambda s: predict(s, 3, trigrams_spa, trigrams_por))
X_test['lan_code_fourgrams'] = X_test['sentence'].apply(lambda s: predict(s, 4, fourgrams_spa, fourgrams_por))
```

### Performance

```python
ngram_performance = pd.DataFrame(columns=['Precision', 'Recall'])
```

```python
from sklearn.metrics import precision_recall_fscore_support

ngram_performance.loc['Unigrams'] = precision_recall_fscore_support(X_test['lan_code'], 
                                                                    X_test['lan_code_unigrams'], 
                                                                    average='micro')[:2]
```

```python
ngram_performance.loc['Bigrams'] = precision_recall_fscore_support(X_test['lan_code'], 
                                                                   X_test['lan_code_bigrams'], 
                                                                   average='micro')[:2]
```

```python
ngram_performance.loc['Trigrams'] = precision_recall_fscore_support(X_test['lan_code'], 
                                                                    X_test['lan_code_trigrams'], 
                                                                    average='micro')[:2]
```

```python
ngram_performance.loc['4grams'] = precision_recall_fscore_support(X_test['lan_code'], 
                                                                  X_test['lan_code_fourgrams'], 
                                                                  average='micro')[:2]
```
